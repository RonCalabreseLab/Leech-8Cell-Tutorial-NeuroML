```python
!pip install numpy
!pip install pandas
!pip install matplotlib
import pandas as pd
import numpy as np

voltage_range = np.arange(-100,+50,150/3001)
```

```python
chan_Na = pd.read_csv("chan_Na.m", header=None, sep="\t")

tau_Na = pd.DataFrame({
   "NA_inact" : 1/chan_Na[3],
   'NA_act' : 1/chan_Na[1],
   'voltage' : voltage_range
   })
tau_Na.set_index('voltage', inplace=True)

minf_Na = pd.DataFrame({
    "NA_act" : chan_Na[0]/chan_Na[1],
    "NA_inact" : chan_Na[2]/chan_Na[3]
    })
minf_Na['voltage'] = voltage_range
minf_Na.set_index('voltage', inplace=True)

tau_Na[{"NA_inact","NA_act"}].plot.line(color={"NA_inact": "green", "NA_act": "blue"}, title="Tau", ylabel="seconds", grid=True) 
minf_Na[{"NA_inact","NA_act"}].plot.line(color={"NA_inact": "green", "NA_act": "blue"}, title="Minf", ylabel="activation", grid=True)
```

```python
chan_P = pd.read_csv("chan_P.m", header=None, sep="\t")
chan_K2 = pd.read_csv("chan_K2.m", header=None, sep="\t")

tau_K2P = pd.DataFrame({
   "K2_act" : 1/chan_K2[1],
   'P_act' : 1/chan_P[1]
   })
tau_K2P['voltage'] = voltage_range
tau_K2P.set_index('voltage', inplace=True)

minf_K2P = pd.DataFrame({
    "K2_act" : chan_K2[0]/chan_K2[1],
    "P_act" : chan_P[0]/chan_P[1]
    })
minf_K2P['voltage'] = voltage_range
minf_K2P.set_index('voltage', inplace=True)

tau_K2P.plot.line(color={"K2_act": "orange", "P_act": "blue"}, title="Tau", ylabel="seconds", grid=True) 
minf_K2P.plot.line(color={"K2_act": "orange", "P_act": "blue"}, title="Minf", ylabel="activation", grid=True)
```
